# Face detection

```python
# import modules
import cv2
import numpy as np
import os 

# our images data captrued here
features = []
# our labeled data captured here
label  =[] 

# taking username as label
name = input("enter name : ")

# open vdo stream
vdo = cv2.VideoCapture(0)
# initialize cascadeclassidier
obj = cv2.CascadeClassifier(r"C:\Users\Aditya\AppData\Local\Programs\Python\Python39\Lib\site-packages\cv2\data//haarcascade_frontalface_default.xml")

# looping for geting each frames from vdo stream 
while True:
    
    flag,frames = vdo.read()
    if flag:
        # detecting face
        face = obj.detectMultiScale(frames)
        # iterate through each face         
        for f in face:
            x,y,w,h = f
            
            # cutting face             
            cutting_face = frames[y:y+h,x:x+w]
            
            # resizing image            
            sample = cv2.resize(cutting_face,(100,100))
            
            # convert image to gray scale
            gray = cv2.cvtColor(sample,cv2.COLOR_BGR2GRAY)
    
            # drawing rectangle in window for make sure face detection              
            cv2.rectangle(frames,(x,y),(x+w,y+h),(255,255,255),3)
        
        cv2.imshow("img",frames)

    key = cv2.waitKey(1) 
    if key==ord('q'):
        break
    # capturing perticular frame by pressing c     
    if key==ord('c'):
        # save our features and lables
        features.append(gray.flatten())
        label.append([name])


# building training data
training_data = np.hstack([np.array(label),np.array(features)])

file_name="face_data.npy"

# if file is already exist then just append new data to it
if os.path.exists(file_name):
    old_data = np.load(file_name)
    training_data= np.vstack([old_data,training_data])
    
# save this whole data in to file so that it stays in our machine after executing this program
np.save(file_name,training_data)

vdo.release()
cv2.destroyAllWindows()
```
