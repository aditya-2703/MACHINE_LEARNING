# Web scraping for fetching quotes

importing lib

```python
from bs4 import BeautifulSoup as bs
import requests as r
```

fetching data from url

```python
url = r.get("https://pakwired.com/100-best-quotes-time/")
```

loading html content

```python
data = bs(url.content)
```

```python
quotes_list = data.find_all("blockquote",attrs={"class":"shortcode_quote"})
author_list = data.find_all("cite")
```

```python
quotes_data = []
for i in range(len(quotes_list)):
    quotes_data.append({
        "quotes":quotes_list[i].p.text,
        "author":author_list[i].text[1:-1].strip()
    })
```
